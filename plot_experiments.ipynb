```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def p_norm(p):
    xx, yy = np.meshgrid(np.linspace(-3, 3, num=101), np.linspace(-3, 3, num=101))
    zz = ((np.abs((xx))**p) + (np.abs((yy))**p))**(1./p)
    return xx, yy, zz
```

## $\ell^1$-norm Minimization

```python
p1 = p_norm(1)
p2 = p_norm(2)

plt.figure(figsize=(10, 10))
plt.grid(True)
plt.contour(p1[0], p1[1], p1[2], [1], colors='red', linewidths=2) 
plt.contour(p2[0], p2[1], p2[2], [1], colors='black', linewidths=2)
plt.contour(p1[0], p1[1], p1[2], [np.sqrt(2)], colors='green', linewidths=2)
plt.quiver([0, 0], [0, 0], [np.sqrt(2)/2, 1], [np.sqrt(2)/2, 0], 
           angles='xy', scale_units='xy', scale=1, color=["green", "red"])
plt.ylim(-2, 2)
plt.xlim(-2 ,2)
plt.show()
```

```python
p4 = p_norm(4)
p2 = p_norm(2)

plt.figure(figsize=(10, 10))
plt.grid(True)
plt.contour(p4[0], p4[1], p4[2], [1], colors='red', linewidths=2) 
plt.contour(p2[0], p2[1], p2[2], [1], colors='black', linewidths=2)
plt.contour(p4[0], p4[1], p4[2], [0.84], colors='green', linewidths=2)
plt.quiver([0, 0], [0, 0], [np.sqrt(2)/2, 1], [np.sqrt(2)/2, 0], 
           angles='xy', scale_units='xy', scale=1, color=["green", "red"])
plt.ylim(-2, 2)
plt.xlim(-2 ,2)
plt.show()
```
